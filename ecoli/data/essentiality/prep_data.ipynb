```python
genes = [g.name for g in ecoli.genes]
genes

gene_mapper = {g.name:g.id for g in ecoli.genes}
# for gene in ecoli.genes:
#     print(gene.name)
```

```python
df = pd.read_csv('Escherichia_coli_K12_genes.csv')
```

```python
df.head()
```

```python
df = df.loc[df.gene.isin(genes)].copy()
```

```python
df['essential'] = False
df.loc[df.essentiality=='E', 'essential'] = True
```

```python
gene_mapper
```

```python
df['gene'] = df.gene.replace(gene_mapper)
```

```python
df[['gene', 'essential']].to_csv('gene_ess.csv', index=False)
```
