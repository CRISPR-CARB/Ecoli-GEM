```python
obs_mapper = pd.read_csv('obs_id_to_growth.txt', delimiter='\t', names=['obs', 'growth'], skiprows=1)
```

```python
obs_mapper.loc[obs_mapper.growth == 'NONE', 'growth'] = False
obs_mapper.loc[obs_mapper.growth == 'LOW', 'growth'] = True
obs_mapper.loc[obs_mapper.growth == 'NORMAL', 'growth'] = True
obs_mapper = obs_mapper.set_index('obs').to_dict()['growth']
```

```python
growth_list = []
for plate, growth in df[['Common-Name', 'Observed-Growth']].values:
#     print(growth)
    obs = growth.split(' // ')
    count = 0
    for o in obs:
        count+=obs_mapper[o]
    fraction_experiments_grow = count/len(obs)
    growth_list.append(fraction_experiments_grow>.5)
growth_list
df['growth'] = growth_list
```
